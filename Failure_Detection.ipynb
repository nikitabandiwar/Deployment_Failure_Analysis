```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Read CSV with NaN values specified
df = pd.read_csv('deployment_detail.csv', na_values=['', 'NA', 'NaN'])
```

```python
import csv

with open('deployment_detail.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    rows = list(csv_reader)
    df.head()
    print(rows[7])
    print(rows[22])
    
```

```python
df =df.dropna(subset=['failure_category', 'errors_message'])
```

```python
pd.set_option('display.max_columns', None)
print(df)
```

```python
# Text preprocessing and modeling pipeline
model = make_pipeline(CountVectorizer(), MultinomialNB())
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['errors_message'], df['failure_category'], random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predictions on the test set
predictions = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f'Model Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, predictions))

print(f"Accuracy : {round(accuracy*100,2)}%")
```
